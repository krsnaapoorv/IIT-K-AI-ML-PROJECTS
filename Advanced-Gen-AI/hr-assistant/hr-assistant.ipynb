````python
# Load Nestle's HR Policy PDF
loader = PyPDFLoader("the_nestle_hr_policy_pdf_2012.pdf")  
documents = loader.load()
````

````python
# Split the documents into chunks
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

docs = text_splitter.split_documents(documents)
````

````python
# Create vector representations using ChromaDB and OpenAI embeddings
embedding = OpenAIEmbeddings()

vectordb = Chroma.from_documents(
    docs,
    embedding=embedding,
    persist_directory="./chroma_db"
)

vectordb.persist()
````

````python
# Build the question-answering system
llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0)
retriever = vectordb.as_retriever()
````

````python
prompt_template = """
You are an AI HR Assistant for Nestlé.

Use ONLY the provided context to answer the question.
If the answer is not found, say "I don't know based on the document."

Context:
{context}

Question:
{question}

Answer:
"""
````

````python
# Chatbot Query
def chatbot(query):
    docs = retriever.get_relevant_documents(query)
    context = "\n\n".join([doc.page_content for doc in docs])
    prompt = prompt_template.format(context=context, question=query)
    response = llm.invoke(prompt)
    return response
````

````python
# Create a Gradio interface
interface = gr.Interface(
    fn=chatbot,
    inputs=gr.Textbox(lines=2, placeholder="Ask HR-related questions..."),
    outputs="text",
    title="Nestlé HR Policy Chatbot",
    description="Ask questions about HR policies and get instant answers."
)

interface.launch()
````
